**[Problem 1] Data acquisition**

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

**[Problem 2] Data combination**

```python
df_target = pd.DataFrame(iris.target, columns=['target'])
df = pd.concat([df, df_target], axis=1)
df
```

**[Problem 4] Investigation of the iris dataset itself**

The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters. Based on the combination of these four features, Fisher developed a linear discriminant model to distinguish the species from each other.

**[Problem 6] Creating a diagram**

```python
import matplotlib.pyplot as plt

x = df["target"].value_counts()
labels = ['Iris setosa', 'Iris virginica','Iris versicolor']
fig, ax = plt.subplots()
ax.pie(x, labels=labels,  autopct='%1.0f%%')
plt.show()
```

```python
x1 = df.iloc[:, 0]
x2 = df.iloc[:, 1]
x3 = df.iloc[:, 2]
x4 = df.iloc[:, 3]

fig, ax = plt.subplots()
labels = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
ax.boxplot((x1, x2, x3, x4), labels=labels)
ax.set_xlabel('Features')
ax.set_ylabel('Observed values')
plt.title("Box Plot")
plt.show()
```

```python
x1 = df.iloc[:, 0]
x2 = df.iloc[:, 1]
x3 = df.iloc[:, 2]
x4 = df.iloc[:, 3]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.violinplot((x1, x2, x3, x4))
ax.set_xlabel('Features')
ax.set_ylabel('Observed values')
plt.title("Voilin Plot")
plt.show()
```

[**Problem 7] Confirmation of the relationship between features**

```python
plt.figure(figsize=(16, 8), dpi=80)
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c='r', label='sepal length vs sepal width')
plt.scatter(df.iloc[:, 0], df.iloc[:, 2], label='sepal length vs petal length')
plt.scatter(df.iloc[:, 0], df.iloc[:, 3], label='sepal length vs petal width')
plt.scatter(df.iloc[:, 1], df.iloc[:, 2], label='sepal width vs petal length')
plt.scatter(df.iloc[:, 1], df.iloc[:, 3], label='sepal width vs petal width')
plt.scatter(df.iloc[:, 2], df.iloc[:, 3], label='petal length vs petal width')
plt.legend()
plt.show()
```
