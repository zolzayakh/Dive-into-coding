**[Problem 1] Creation of random numbers**

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
mean = (-3, 0)
cov = np.array([[1, 0.8], [0.8, 1]])
X = np.random.multivariate_normal(mean, cov, 500)

print(X.shape)
print(X)

##[[-2.50340750e+00  2.08656738e-01]
#  [-3.58382140e+00  2.74225346e-02]

#After use the np.random.seed()
# [-4.80006782e+00 -1.54698618e+00]
#  [-4.63714503e+00 -2.19879728e-01]
```

**[Problem 2] Visualization by scatter plot**

```python
np.random.seed(0)
mean = (-3, 0)
cov = [[1, 0.8], [0.8, 1]]
x1, x2 = np.random.multivariate_normal(mean, cov, 500).T
plt.scatter(x1, x2, alpha=0.5, marker='*')
plt.title("Scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

**[Problem 3] Visualization by histogram**

```python
np.random.seed(0)
mean = (-3, 0)
cov = [[1, 0.8], [0.8, 1]]
x1, x2 = np.random.multivariate_normal(mean, cov, 500).T
fig, ax = plt.subplots()
ax.hist(x1)
plt.title("histogram of x1")
plt.xlabel("x1")
plt.ylabel("frequency")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(x2)
plt.title("histogram of x2")
plt.xlabel("x2")
plt.ylabel("frequency")
plt.show()
```

**[Problem 4] Addition of data**

```python
import numpy as np
import matplotlib.pyplot as plt

#prob1
np.random.seed(0)
mean = (0, -3)
cov = np.array([[1, 0.8], [0.8, 1]])
x1, x2 = np.random.multivariate_normal(mean, cov, 500).T
#prob4
mean1 = (-3, 0)
cov1 = np.array([[1, 0.8], [0.8, 1]])
y1, y2 = np.random.multivariate_normal(mean1, cov1, 500).T

plt.scatter(x1, x2, label="0")
plt.scatter(y1, y2, label="1")
plt.title("Scatter")
plt.legend()
plt.xlabel("x1")
plt.ylabel("x2")

plt.show()

```

**[Problem 5] Data combination**

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
#prob1
mean = (0, -3)
cov = np.array([[1, 0.8], [0.8, 1]])
X = np.random.multivariate_normal(mean, cov, 500)
#prob4
mean1 = (-3, 0)
cov1 = np.array([[1, 0.8], [0.8, 1]])
Y = np.random.multivariate_normal(mean1, cov1, 500)

merged_conc = np.concatenate([X, Y], axis=0)
print(merged_conc.shape)
```

```python
merged_vstack = np.vstack((X, Y))
print(merged_vstack.shape)
```
